*Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.*


```python
#!pip3 install translate
#!pip3 install googletrans==4.0.0rc1
text = "Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày."
```

```python
from translate import Translator
translator_en = Translator(to_lang ="en", from_lang="vi")
translator_ja = Translator(to_lang ="ja", from_lang="vi")
translation_en = translator_en.translate(text)
translation_ja = translator_ja.translate(text)
print(translation_en)
print(translation_ja)
```

```python
from googletrans import Translator as tr
tr = tr()
g_translation_en = tr.translate(text, dest = 'en', src ='vi')
g_translation_ja = tr.translate(text, dest = 'ja', src ='vi')
print(g_translation_en.text)
print(g_translation_ja.text)

```
