```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('sqlite2.db')
cur = con.cursor()
```

```python
cur.execute("""
CREATE TABLE STOCK (
    ID INTEGER PRIMARY KEY,
    NAME TEXT NOT NULL,
    BUY INTERGER NOT NULL,
    INVESTOR TEXT NOT NULL
);
""")
```

```python
cur.execute("""
INSERT INTO STOCK(ID, NAME, BUY, INVESTOR)
VALUES
    ('1', 'ACB', '29.45', 'Nguyen'),
    ('2', 'VIC', '44.55', 'Nguyen'),
    ('3', 'GMD', '74.30', 'Nguyen'),
    ('4', 'ACB', '28.45', 'Vinh'),
    ('5', 'VIC', '40.55', 'Vinh'),
    ('6', 'GMD', '60.30', 'Vinh')
""")
con.commit()
```

```python
data = pd.read_sql_query("SELECT * FROM STOCK", con)
print(data)
```

```python
# Cu 1: In ra kt qu thng k tng gi mua (BUY) ca bng STOCK:
query = """
SELECT SUM(BUY) AS TOTAL_BUY
FROM STOCK;
"""
data_sum = pd.read_sql_query(query, con)
print(data_sum)
```

```python
# Cu 2: Thng k m c phiu c gi mua (BUY) ln nht theo nh u t (INVESTOR):
query = """
SELECT INVESTOR, NAME, MAX(BUY) AS MAX_PRICE
FROM STOCK
GROUP BY INVESTOR;
"""
data_max = pd.read_sql_query(query, con)
print(data_max)
```
