```python
import cv2
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# Load image
img = cv2.imread('/Users/chaupham/Data_Science/40-Day-Python/Day-34/dataset/2.jpg',0)
# Apply Gaussian Filter with kernel matrix 5x5, sigma = 2.5, 5.0, 10.0
img_filtered1 = cv2.GaussianBlur(img, (5, 5), 2.5)
img_filtered2 = cv2.GaussianBlur(img, (5, 5), 5.0)
img_filtered3 = cv2.GaussianBlur(img, (5, 5), 10.0)
fig,axis = plt.subplots(2,2)
axis[0,0].imshow(img, cmap ='gray')
axis[0,0].set_title('original image')
axis[0,1].imshow(img_filtered1, cmap ='gray')
axis[0,1].set_title('sigma = 2.5')
axis[1,0].imshow(img_filtered2, cmap ='gray')
axis[1,0].set_title('sigma = 5.0')
axis[1,1].imshow(img_filtered3, cmap ='gray')
axis[1,1].set_title('sigma = 10.0')
```

```python
def gaussian_kernel(size, sigma):
    if size % 2 == 0:
        size = size + 1

    max_point = size // 2 # both directions (x,y) maximum cell start point
    min_point = -max_point # both directions (x,y) minimum cell start point
    K = np.zeros((size, size)) # kernel matrix
    for x in range(min_point, max_point + 1):
        for y in range(min_point, max_point + 1):
            value = (1/(2*math.pi*sigma**2))*math.exp((x**2+y**2)/(-2*sigma**2))
            K[x - min_point, y - min_point] = value
    return K
kernel1 = gaussian_kernel(5, 2.5)
img_gaussian1 = cv2.filter2D(img, -1, kernel1)
kernel2 = gaussian_kernel(5, 5.0)
img_gaussian2 = cv2.filter2D(img, -1, kernel2)
kernel3 = gaussian_kernel(5, 10.0)
img_gaussian3 = cv2.filter2D(img, -1, kernel3)
fig,axis = plt.subplots(2,2)
axis[0,0].imshow(img, cmap ='gray')
axis[0,0].set_title('original image')
axis[0,1].imshow(img_gaussian1, cmap ='gray')
axis[0,1].set_title('sigma = 2.5')
axis[1,0].imshow(img_gaussian2, cmap ='gray')
axis[1,0].set_title('sigma = 5.0')
axis[1,1].imshow(img_gaussian3, cmap ='gray')
axis[1,1].set_title('sigma = 10.0')
```
