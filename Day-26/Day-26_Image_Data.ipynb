#### Câu 2:
Chuyển ảnh thành màu xám và hiển thị

```python
new_img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(new_img2, cmap='gray')
plt.show()
```

#### Câu 3:
Tăng cường độ sáng ảnh màu lên 50 đơn vị, và giảm cường độ sáng xuống 80 đơn vị (trên ảnh màu gốc ban đầu) và hiển thị

```python
# Tăng cường độ sáng ảnh màu lên 50 đơn vị
new_img3 = img.astype(np.float32) + 50
new_img3 = np.clip(new_img3, 0, 255)
new_img3 = new_img3.astype(np.uint8)
plt.imshow(new_img3)
plt.show()
```

```python
# Giảm cường độ sáng ảnh màu xuống 80 đơn vị
new_img4 = img.astype(np.float32) -80
new_img4 = np.clip(new_img4, 0, 255)
new_img4 = new_img4.astype(np.uint8)
plt.imshow(new_img4)
plt.show()
```
