```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# Viết chương trình sử dụng hàm zeros tạo Numpy array, Tensorflow tensor, Pytorch tensor chỉ chứa giá trị là số 0 với kích thước (3,4)
arr_zeros = np.zeros((3,4))
print(arr_zeros)
torch_tensor_zeros = torch.zeros((3,4))
print(torch_tensor_zeros)
tf_tensor_zeros = tf.zeros((3,4))
print(tf_tensor_zeros)
```

```python
# Viết chương trình sử dụng hàm zeros tạo Numpy array, Tensorflow tensor, Pytorch tensor chỉ chứa giá trị là số 1 với kích thước (3,4)
arr_ones = np.ones((3,4))
print(arr_ones)
torch_tensor_ones = torch.ones((3,4))
print(torch_tensor_ones)
tf_tensor_ones = tf.ones((3,4))
print(tf_tensor_ones)
```

```python
# # Viết chương trình sử dụng hàm zeros tạo Numpy array, Tensorflow tensor, Pytorch tensor chỉ chứa giá trị là số 5 với kích thước (3,4)
arr_full = np.full((3,4),5)
print(arr_full)
torch_tensor_full = torch.full((3,4),5)
print(torch_tensor_full)
tf_tensor_fill = tf.fill((3,4),5)
print(tf_tensor_fill)
```
